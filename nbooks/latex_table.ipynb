# Make LaTex Table

```python
import csv
import numpy as np
```

## Load Table Data

```python
with open('../tables/table2.csv', 'r') as csvfile:
    reader = csv.reader(csvfile, delimiter=",")
    data = np.array([row for row in reader]).T

obsnames = data[0]
z_list, fluxpeaks, fluxpeaks_err, ul_list, m1450_list, allwise_AB_list, allwise_err_list, spitzer_AB_list, spitzer_err_list, Lsgn, log_L_nu_list, log_L_peak_list, log_L_peak_err_list, log_L_ul_list, log_L4400_list, log_L4400_err_list, R_list, Rerr_list, Rul_list = data[1:].astype(float)
```

```python
i_det, i_ul = [], []

for i, (fp, fperr) in enumerate(zip(fluxpeaks, fluxpeaks_err)):
    if fp/fperr > 3:
        i_det.append(i)
    else:
        i_ul.append(i)
```

```python
det_str = "{:s} & {:.2f} & & {:.0f}$\pm${:.0f} & {:.1f} & {:s} & {:s} & {:.2f}$\pm${:.2f} & {:.2f}$\pm${:.2f} & {:.1f}$\pm${:.1f} \\\\"
ul_str = "{:s} & {:.2f} & & <{:.0f} & {:.1f} & {:s} & {:s} & <{:.2f} & {:.2f}$\pm${:.2f} & <{:.1f} \\\\"

for i in range(len(obsnames)):
    name = obsnames[i][4:]

    if np.isnan(spitzer_AB_list[i]):
        spitzer_str = "-"
    else:
        spitzer_str = "{:.2f} $\pm$ {:.2f}".format(spitzer_AB_list[i], spitzer_err_list[i])

    if np.isnan(allwise_AB_list[i]):
        allwise_str = "-"
    else:
        allwise_str = "{:.2f} $\pm$ {:.2f}".format(allwise_AB_list[i], allwise_err_list[i])

    if i in i_det:
        row_str = det_str.format(name, z_list[i], fluxpeaks[i], fluxpeaks_err[i], m1450_list[i], allwise_str, spitzer_str, log_L_peak_list[i], log_L_peak_err_list[i], log_L4400_list[i], log_L4400_err_list[i], R_list[i], Rerr_list[i])
    elif i in i_ul:
        row_str = ul_str.format(name, z_list[i], ul_list[i], m1450_list[i], allwise_str, spitzer_str, log_L_ul_list[i], log_L4400_list[i], log_L4400_err_list[i], Rul_list[i])
    
    print(row_str)
```
